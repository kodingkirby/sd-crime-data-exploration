```python
pd_dict_df = pd.read_csv(r'https://seshat.datasd.org/pd/pd_calls_for_service_dictionary_datasd.csv')
print(pd_dict_df)
```

```python
dispo_codes_df = pd.read_csv(r'https://seshat.datasd.org/pd/pd_dispo_codes_datasd.csv')
print(dispo_codes_df)
```

```python
call_type_df = pd.read_csv(r'https://seshat.datasd.org/pd/pd_cfs_calltypes_datasd.csv')
print(call_type_df)
```

```python
df2 = df.query("call_type != 'NaN'")
```

```python
call_type_df.head(5)
```

```python
call_type_df2 = call_type_df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'])
```

```python
call_type_df2.head(5)
```

```python
df2.head(5)
```

```python
calls_df = pd.merge(df2,
                    call_type_df2,
                    left_on='call_type',
                    right_on='call_type',
                    how='left')
```

```python
calls_df.head(5)
```

```python
calls_df2 = pd.merge(calls_df,
                     dispo_codes_df,
                     left_on='disposition',
                     right_on='dispo_code',
                     how='left')
```
